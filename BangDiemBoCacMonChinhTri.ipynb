```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
#load data
pima=pd.read_excel("BangDiem/bang_diem_tong_hop.xlsx")
```

```python
pima
```

```python
pima=pima.drop(columns=['MSSV','Ngày Sinh','diemtbclan1','diemtbccaonhat','Trung bình cộng học tập', 'Trung bình cộng tích lũy','tongsomonconno','tongsohoctrinhconno','Số tín chỉ tích lũy','Điểm rèn luyện'])
pima
```

```python
pima=pima.drop(columns=['Triết học Mác- Lê Nin F1','Những nguyên lý cơ bản của chủ nghĩa Mác- Lê Nin F1','Chủ nghĩa xã hội khoa học','Những nguyên lý cơ bản của chủ nghĩa Mác- Lê Nin F2','Kinh tế chính trị Mác-Lênin','Tư tưởng Hồ Chí Minh','Lịch sử Đảng Cộng sản Việt Nam','Đường lối cách mạng Đảng Cộng sản Việt Nam'])
pima
```

```python
for index, row in pima.iterrows():
    # Duyệt qua từng ô trong hàng
    for column in pima.columns:
        # Kiểm tra xem có dấu phẩy giữa hai giá trị không
        if ',' in str(row[column]):
            # Tách giá trị thành hai phần
            values = str(row[column]).split(',')
            # Chuyển đổi các giá trị thành số nguyên
            values = [float(value) if value != '' else np.nan for value in values]
            # Lấy giá trị lớn hơn
            max_value = max(values)
            # Cập nhật lại giá trị trong ô thành giá trị lớn hơn
            pima.at[index, column] = max_value
```

```python
pima
```

```python
pima=pima.drop([0,1])
pima=pima.fillna(0)
```

```python
pima.to_excel('BangDiem/bang_diem_clean.xlsx',index=False)
```

```python
df=pima.copy()
X=df.drop('Xếp loại học tập',axis=1)
Y= df['Xếp loại học tập']
```

```python
#%pip install scikit-learn 
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 90% training and 10% test
from sklearn import tree
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
# Build a Gaussian Classifier
model_naive = GaussianNB()
model_naive.fit(X_train, y_train)
```

```python
y_pred_naive=model_naive.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_naive))
```
