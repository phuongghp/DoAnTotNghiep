```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
#load data
pima=pd.read_excel("BangDiem/diem_trung_binh.xlsx")
```

```python
pima
```

```python
pima=pima.drop(columns=['mssv'])
pima
```

```python
df=pima.copy()
X=df.drop('xep_loai_hoc_tap',axis=1)
Y= df['xep_loai_hoc_tap']
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 90% training and 10% test
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
# Build a Gaussian Classifier
model_naive = GaussianNB()
model_naive.fit(X_train, y_train)
```

```python
y_pred_naive=model_naive.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_naive))
```
