```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
#load data
pima=pd.read_excel("BangDiem/SinhVienNam1_HeChu_clean.xlsx")
```

```python
pima
```

```python
for index, row in pima.iterrows():
    for col in pima.columns:
        cell_value = row[col]
        if isinstance(cell_value, (str)) and cell_value =='A':
            pima.at[index, col] = 1
        if isinstance(cell_value, (str)) and cell_value =='B':
            pima.at[index, col] = 2
        if isinstance(cell_value, (str)) and cell_value =='C':
            pima.at[index, col] = 3
        if isinstance(cell_value, (str)) and cell_value =='D':
            pima.at[index, col] = 4
        if isinstance(cell_value, (str)) and cell_value =='F':
            pima.at[index, col] = 5
```

```python
pima
```

```python
df=pima.copy()
X=df.drop('Xếp loại học tập',axis=1)
Y= df['Xếp loại học tập']
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 90% training and 10% test
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
# Build a Gaussian Classifier
model_naive = GaussianNB()
model_naive.fit(X_train, y_train)
```

```python
y_pred_naive=model_naive.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_naive))
```

```python
import pickle
#pickle.dump(model,open('Pickle/sinhvienn1_heso.pkl','wb'))
pickle.dump(model_naive,open('Pickle/sinhvienn1_naive_hechu.pkl','wb'))
```
